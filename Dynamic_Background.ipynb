```python
from rembg import remove
import requests
from PIL import Image
from io import BytesIO
import os
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
input_img = '/content/drive/MyDrive/p.jpeg'
output_path='/content/drive/MyDrive/outputIMG/pic.jpeg'
with open(output_path,'wb') as f:
  input=open(input_img,'rb').read()
  subject=remove(input,alpha_matting=True,alpha_matting_foreground_threshold=50)
  f.write(subject)
```

```python
image_path=input_img
image_size=Image.open(image_path)
background_img='/content/drive/MyDrive/background.jpeg'
background_img=Image.open(background_img)
background=background_img.resize((image_size.width,image_size.height))
```

```python
foreground=Image.open(output_path)
background.paste(foreground,(0,0),foreground)
display(background)
```
